```python
@dataclass(frozen=True)
class Runner:
    id: int
    category: int
    name: str
    time: Optional[str]  # None = no time
```

## Parsen der Anmeldungen


```python
def parse_runners_df(text: str) -> pd.DataFrame:
    """Parses registration text robustly: id, category, and full name (incl. numbers)."""
    rows = []
    for line in text.strip().split("\n"):
        parts = line.strip().split("\t")
        if len(parts) < 3:
            continue  # Skip invalid lines
        try:
            runner_id = int(parts[0])
            category = int(parts[1])
            name = " ".join(parts[2:]).strip()  # Keep numbers like "Tobler 2"
            rows.append((runner_id, category, name))
        except ValueError:
            print(f"Skipping line due to parsing error: {line}")
    df = pd.DataFrame(rows, columns=["id", "category", "name"])
    df["time"] = None
    return df
```

## Finish-Zeiten aus Text extrahieren (ID + Zeit -> DataFrame)

```python
def parse_finish_times_df(text: str) -> pd.DataFrame:
    """Parses finish times into a DataFrame with id and time."""
    rows = []
    for line in text.strip().split("\n"):
        parts = line.strip().split("\t")
        if len(parts) == 2:
            try:
                rows.append((int(parts[0]), parts[1]))
            except ValueError:
                print(f"Skipping line due to parsing error: {line}")
    return pd.DataFrame(rows, columns=["id", "time"])
```

## Finish-Zeiten den Läufern zuweisen (mit ID-Merge)

```python
def assign_times_df(runners_df: pd.DataFrame, finish_times_df: pd.DataFrame) -> pd.DataFrame:
    """Merges runners with finish times."""
    merged_df = runners_df.merge(finish_times_df, on="id", how="left", suffixes=('', '_new'))
    merged_df["time"] = merged_df["time_new"].combine_first(merged_df["time"])
    return merged_df.drop(columns=["time_new"])
```

## Läufer sortieren nach Zeit oder Kategorie + Zeit

```python
def sort_runners_df(df: pd.DataFrame, sort_by: str) -> pd.DataFrame:
    """Sorts runners by time or category + time."""
    def parse_time(t: Optional[str]) -> datetime.datetime:
        try:
            return datetime.datetime.strptime(t, "%H:%M:%S")
        except:
            return datetime.datetime.max

    df["time_sort"] = df["time"].apply(parse_time)

    if sort_by == "time only":
        sorted_df = df.sort_values(by="time_sort")
    elif sort_by == "category first":
        sorted_df = df.sort_values(by=["category", "time_sort"])
    else:
        raise ValueError("Invalid sort option.")

    return sorted_df.drop(columns=["time_sort"])
```

# Abfrage der Sortieroption und Ausgabe der sortierten Läuferliste

```python
def ask_sorting_preference() -> str:
    choice = input("Sort by (1) Time only or (2) Category first, then Time? ")
    if choice == "1":
        return "time only"
    elif choice == "2":
        return "category first"
    else:
        print("Invalid input. Please choose 1 or 2.")
        return ask_sorting_preference()

def display_results_df(df: pd.DataFrame):
    df_display = df.copy()
    df_display["time"] = df_display["time"].fillna("no time")
    print(df_display[["id", "category", "name", "time"]].to_string(index=False))
```
